Geometric function:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_image(size=(100, 100)):
    return np.zeros((size[0], size[1], 3), dtype=np.uint8)

def draw_rectangle(image, top_left, bottom_right, color=(255, 255, 255)):
    image[top_left[1]:bottom_right[1], top_left[0]:bottom_right[0]] = color
```

```python
def translate(image, dx, dy):
    translated_image = np.zeros_like(image)
    rows, cols, _ = image.shape
    translated_image[max(0, dy):min(rows + dy, rows), max(0, dx):min(cols + dx, cols)] = image[max(0, -dy):min(rows - dy, rows), max(0, -dx):min(cols - dx, cols)]
    return translated_image
```

```python
image_size = (200, 200)
image = create_image(size=image_size)

# Draw a white rectangle on the image
draw_rectangle(image, (50, 50), (150, 150), color=(255, 255, 255))

# Display the original image
plt.imshow(image)
plt.title("Original Image")
plt.show()
```

```python
# Translate the image
translated_image = translate(image, dx=20, dy=-30)

# Display the translated image
plt.imshow(translated_image)
plt.title("Translated Image")
plt.show()
```

Arithmetic function:

```python
def create_image(size=(200, 200), square_size=50):
    image = np.ones((size[0], size[1], 3), dtype=np.uint8) * 255  # White background
    square_start = (size[0] // 2 - square_size // 2, size[1] // 2 - square_size // 2)
    square_end = (square_start[0] + square_size, square_start[1] + square_size)
    image[square_start[0]:square_end[0], square_start[1]:square_end[1]] = [0, 0, 0]  # Black square
    return image
```

```python
def add_images(image1, image2):
    return np.clip(image1 + image2, 0, 255).astype(np.uint8)

# Create two images with black squares
image1 = create_image()
image2 = create_image()
```

```python
# Display the original images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.title("Image 1")

plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.title("Image 2")

plt.show()
```

```python
# Add the two images
result_image = add_images(image1, image2)

# Display the result image
plt.imshow(result_image)
plt.title("Result Image (Addition)")
plt.show()
```

```python
def translate_image(image, dx, dy):

    rows, cols, channels = image.shape

    # Create an empty image for the translated result
    translated_image = np.zeros_like(image, dtype=np.uint8)

    # Apply translation manually
    for i in range(rows):
        for j in range(cols):
            new_i = i + dy
            new_j = j + dx

            # Check if the new indices are within bounds
            if 0 <= new_i < rows and 0 <= new_j < cols:
                translated_image[new_i, new_j] = image[i, j]

    return translated_image
```

```python
# Define the size of the image in rows and columns
rows = 100
cols = 100

# Create a 2D NumPy array with a diagonal line separating white and black pixels
image = np.zeros((rows, cols, 3), dtype=np.uint8)

# Set the upper part of the diagonal to white (255, 255, 255) and the lower part to black (0, 0, 0)
for i in range(rows):
    for j in range(cols):
        if i < j:
            image[i, j] = [255, 255, 255]  # White pixel
        else:
            image[i, j] = [0, 0, 0]  # Black pixel

# Print the array
print(image)

# Display the image using Matplotlib
plt.imshow(image)
plt.axis('on')
plt.show()
```

```python
rows = 2
cols = 2

# Create an array of 1 and 0 alternately
array = np.zeros((rows, cols), dtype=int)

for i in range(rows):
    for j in range(cols):
        if (i + j) % 2 == 1:
            array[i, j] = 1

# Display the array
print("Array:")
print(array)

# Display the chessboard pattern
plt.imshow(array, cmap='gray', interpolation='nearest')
plt.axis('on')
plt.show()
```

```python
# Define the size of the chessboard in rows and columns
rows = 8
cols = 8

# Create an array of 1 and 0 alternately
array = np.zeros((rows, cols), dtype=int)

for i in range(rows):
    for j in range(cols):
        if (i + j) % 2 == 1:
            array[i, j] = 1

# Display the array
print("Array:")
print(array)

# Display the chessboard pattern
plt.imshow(array, cmap='gray', interpolation='nearest', extent=(0, cols + 0, rows + 0, 0))
plt.axis('on')
plt.show()
```

Pattern of D8 and De

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid size
gridSize = 20

# Generate grid coordinates
x, y = np.meshgrid(np.arange(1, gridSize + 1), np.arange(1, gridSize + 1))

# Define the center point
center = [gridSize / 2, gridSize / 2]

# Calculate D8 distance
d8_distance = np.maximum(np.abs(x - center[0]), np.abs(y - center[1]))

# Calculate Euclidean distance
euclidean_distance = np.sqrt((x - center[0]) ** 2 + (y - center[1]) ** 2)

# Plot D8 distance
plt.subplot(1, 2, 1)
plt.imshow(d8_distance, cmap='gray')  # Set the colormap to grayscale
plt.title('D8 Distance')
plt.colorbar()
plt.gca().set_aspect('equal', adjustable='box')  # Set aspect ratio to make it square
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plot Euclidean distance
plt.subplot(1, 2, 2)
plt.imshow(euclidean_distance, cmap='gray')  # Set the colormap to grayscale
plt.title('Euclidean Distance')
plt.colorbar()
plt.gca().set_aspect('equal', adjustable='box')  # Set aspect ratio to make it square
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
